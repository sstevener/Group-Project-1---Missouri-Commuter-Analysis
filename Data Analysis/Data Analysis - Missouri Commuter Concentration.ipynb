```python
#import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gmaps
from config import gkey

#set gmaps settings
gmaps.configure(api_key = gkey)

#import MO commuter & non-MO metro commuter csv
county_commute_data = pd.read_csv("..\Resources\county_commute_data.csv")
other_state_commute_data = pd.read_csv("..\Resources\other_state_commute_data.csv")
```

```python
all_county_commute_data = county_commute_data.append(other_state_commute_data)
all_county_commute_data
```

```python
##plot heatmap of # of commuters
#set gmaps figure
all_county_commuter_fig = gmaps.figure()

#set coordinates as locations
locations = all_county_commute_data[["Latitude", "Longitude"]]

#plot heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=all_county_commute_data["Total Commuters"], 
                                 dissipating=False, max_intensity=max(all_county_commute_data["Total Commuters"]),
                                 point_radius = .333)

#adjust heat layer setting to help with heatmap dissipating on zoom
heat_layer.dissipating = False
heat_layer.max_intensity = max(all_county_commute_data["Total Commuters"])
heat_layer.point_radius = .333
#add heat layer to figure
all_county_commuter_fig.add_layer(heat_layer)

all_county_commuter_fig
```
